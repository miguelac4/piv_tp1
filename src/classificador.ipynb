```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from ipywidgets import interactive_output, HBox, VBox
import math
from ipywidgets import FileUpload
import os
import time

# bot√£o de upload
uploader = FileUpload(accept='', multiple=False)
display(uploader)

# espera at√© o ficheiro ser carregado
print("üïê √Ä espera de upload...")
while not uploader.value:
    time.sleep(0.5)  # pequena pausa para n√£o bloquear o notebook

# extrair conte√∫do do upload
ficheiro = next(iter(uploader.value.values()))
conteudo = ficheiro['content']
nome = ficheiro['metadata']['name']

# converter bytes ‚Üí imagem
img = cv2.imdecode(np.frombuffer(conteudo, np.uint8), cv2.IMREAD_COLOR)

print(f"‚úÖ Ficheiro '{nome}' carregado com sucesso!")

# mostrar imagem
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
